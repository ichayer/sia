```python
from tp4.Oja.pca_plots import *
from tp4.tools import csv_to_dict, standardize_data

countries, labels, country_data = csv_to_dict("./europe.csv")
standardized_country_data = standardize_data(country_data)

labels.pop(0) # Drop country label

standardized_matrix = np.array(list(standardized_country_data.values()), dtype=np.float64)
not_standardized_matrix = np.array(list(country_data.values()), dtype=np.float64)
```

```python
plot_boxplot(not_standardized_matrix, 'Boxplot con datos no estandarizados', labels)
# Podemos observar que con los datos NO estandarizados, las caracter√≠sticas Area y GDP en principio parecerian
# ser los que mas condicionarian a comparacion del resto.
```

```python
plot_boxplot(standardized_matrix, 'Boxplot con datos estandarizados', labels)
# Sin embargo, al estandarizar los datos, se nota claramente la variabilidad de dichas caracter√≠sticas
# permitiendonos compararlas entre si.
```

```python
plot_biplot_with_sklearn(standardized_matrix, countries, labels)
# La PC1 separa las caracteristicas en dos:
# GDP, Pop.growth y Life expect
# Military, Unemployment, Area e Inflation

# PC2 a su vez subdivide a los dos grupos de PC1 en:
# GDP
# Life expect y pop growth
# Military y Unemployment
# Area e Inflation
# Resultando asi 4 cuadrantes.

# Vemos que Ucrania es el pais con mayor inflacion anual
# Grecia es el pais que mas presupuesto militar tiene
# Croacia es el pais con mayor tasa de desmpleo
# Luxemburgo es el pais con mas producto bruto interno.
```

```python
plot_PCA1_barchart_with_sklearn(standardized_matrix, countries, labels)
# Si PC1 > 0, el pais se ve mas afectado por Area, Inflation, Uneemployment y Military
# Si PC1 < 0, el pais se ve mas afectado por GDP, Pop.growth y Life expect
```
